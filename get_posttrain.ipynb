```python
for j in range(1100,1166):
    lst = []
    with open ('./data/pubmed23n'+str(j)+'.xml','r',encoding = 'utf-8') as f:
        for i in f.readlines():
            i = i.strip()
            if i[1:13] == 'AbstractText':
                a = i[14:-15]
                lst.append(a)

    print(len(lst))

    with open('posttrain.txt','a') as f:
        for i in range(len(lst)):
            tmp = lst[i].split(' ')
            l = len(tmp)
            if l < 128:
                continue
            f.write(lst[i]+'\n')
```
